```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('Кинетика.csv', delimiter=';')
data
```

```python
x = data['t.min'].head(180)
y = data['[A]exp'].head(180)
fig = px.scatter(data_frame=data, x=x, y=y, title='[A](t) для вычисления начальной скорости 1')

#Построение линии тренда с фиксированным свободным членом равным начальной концентрации
coeffs = np.polyfit(x, y, deg=1)
coeffs[1] = data['[A]exp'].values[0]
poly_eq = np.poly1d(coeffs)
x_trend = np.linspace(x.min(), x.max(), 100)
y_trend = poly_eq(x_trend)
fig.add_scatter(x=x_trend, y=y_trend, mode='lines', name='Линия тренда', line=dict(color='red'))

y_pred = poly_eq(x) 
r_squared = r2_score(y, y_pred)


print('Коэффициенты уравнения линии тренда: a = ' + str(coeffs[0])+', b = '+str(coeffs[1]))
print('R^2: '+str(r_squared))

fig.show()
r1 = -(coeffs[0])
print('r1 = '+str(r1))
```
