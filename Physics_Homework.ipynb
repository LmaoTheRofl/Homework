```python
# Вариант 5
# 8 номер
# R = 1 мм
# P = 15 Вт
# l = 40 мм
c = 3*(10**8) # м/с
P = 15 # Вт
l = 40*(10**(-3)) # м
R = 1*(10**(-3)) # м
# Pressure = (2*P/(S*c))*(np.cos(teta)**2) давление света под углом teta
# где S = 4*pi(l**2+x**2) площадь поверхности сферы
def func(x):
    return ((P*(l**2))/(4*np.pi*c*((l**2)+(x**2))**2))*2*np.pi*x
Pressure = quad(func, 0, R) # интеграл от 0 до R
print('Давление света равно '+str(Pressure[0])+' Па')
```

```python
# Вариант 5
# 13 номер
# m = 2.73E-30 кг
# E = 2.75 эВ
# l = 5.6E-11 м
# U0 = 4.00 эВ
# U(x) = U0(1-x^2/l^2)
# D = ?
h = 6.62*(10**(-34)) # постоянная Планка Дж*с
c = 3*(10**8) # м/с
q = 1.6*10**(-19) #Кл
me = 9.11*(10**(-31)) # кг масса электрона
R = 1.097*(10**(7)) # 1/м Постоянная Ридберга
hp = (h/(2*np.pi)) # приведенная постоянная Планка

m = (2.73*(10**(-30)))
E = 2.75*q # Дж
l = 5.6*(10**(-11))
U0 = 4.00*q # Дж
# U(x) = E
# U(x) = U0*(1-x^2/l^2) = E
# => x1,x2 = l*sqrt(1-E/U0)
x2 = l*np.sqrt(1-(E/U0))
# D = exp[(-2/hp)*2*np.sqrt(2*m*l)*Int[0, x2](sqrt(U(x) - E)dx)] коэффициент прозрачности
def func(x):
    return np.sqrt(U0 - (((x**2)/(l**2))*U0) - E)
I = quad(func, 0, x2) # интеграл от 0 до x2 который затем удваивается т.к. парабола симметричная

D = np.exp((-2/hp)*2*np.sqrt(2*m)*I[0])
print(D)

```
